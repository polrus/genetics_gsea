```python
gsea_disease = pd.read_csv("/Users/polina/genetics_gsea/data/gsea/from_database/disease_zscore_reactome_2025.tsv", sep="\t")
```

```python
gsea_disease
```

```python
def process_gsea_disease(df: pd.DataFrame) -> pd.DataFrame:
    # Step 1: filter
    df = df[df["fdr"] <= 0.05].copy()

    # Step 2: explode propagated_edge
    df["approvedSymbol"] = df["propagated_edge"].str.split(",")
    df = df.explode("approvedSymbol")
    df["approvedSymbol"] = df["approvedSymbol"].str.strip()

    # Step 3: split positive / negative
    positive = df[df["nes"] > 0].groupby("approvedSymbol").agg(
        positiveTerm=("Term", lambda x: ";".join(map(str, x))),
        positiveID=("ID", lambda x: ";".join(map(str, x)))
    )

    negative = df[df["nes"] < 0].groupby("approvedSymbol").agg(
        negativeTerm=("Term", lambda x: ";".join(map(str, x))),
        negativeID=("ID", lambda x: ";".join(map(str, x)))
    )

    # Step 4: merge results
    result = pd.merge(positive, negative, on="approvedSymbol", how="outer").reset_index()

    return result
```

```python
gsea_disease_proc = process_gsea_disease(gsea_disease)
```

```python
gsea_disease_proc
```

```python
geneset_annotated = pd.merge(geneset, gsea_disease_proc, on="approvedSymbol", how="left")
```

```python
geneset_annotated.to_csv("/Users/polina/genetics_gsea/data/geneset_annotated/geneset_reactome_2025_disease_v1.csv", index=False)
```
